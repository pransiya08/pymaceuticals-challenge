### Analysis

- All mice started with a tumor volume of 45mm3 but as the study progressed, we could see that some drugs were effective in shrinking the tumor, while others seemed to slow it down, and the rest were just not effective in stopping the tumor growth.

- As seen from the summary statistics table, Capomulin and Ramicane are the 2 most effective drugs in shrinking the tumors in the mice. This can also be due to the fact that there are more mice tested for these 2 drug regimens, as compared to the others.

- Although, as compared to Capomulin, it can be claimed that Ramicane is more effective given the lower range of the final tumor volume, as well as the less dispersed final tumor volume results compared to that of Capomulin. These can be interpreted as, generally speaking, Ramicane shrunk the tumors to sizes smaller than Capomulin can. Also, The final tumor volume results are more consistent to each other as compared to those of Capomulin's.

- Even though the smallest final tumor volume is from a mouse treated with Infubinol, it is proven to be misleading on its efficacy as this is just an outlier from the data distribution for this drug regimen group. Most of the mice treated with Infubinol still manifested growth in their tumors.

- In interpreting the scatter plots, we could see a direct relationship between the mouse weight and its average tumor size, which means that the higher the weight, the larger the tumor size is. Having a correlation coefficient of 0.84, which is close to the value of 1, actually supports this inference.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

import numpy as np
from scipy.stats import sem
import random
from scipy import stats

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
pyma = pd.merge(study_results,mouse_metadata,how="left",on="Mouse ID")

# Display the data table for preview
pyma
```

```python
# Checking the number of mice.
num_mice = pyma['Mouse ID'].value_counts().count()
num_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_mouse = pyma.loc[pyma[["Mouse ID","Timepoint"]].duplicated() == True]
dup_mouse["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dupe_mouse = pyma.loc[pyma["Mouse ID"] == "g989"]
dupe_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
drop_index = pyma.loc[pyma["Mouse ID"] == "g989"].index
no_dupe_pyma = pyma.drop(index=drop_index)
no_dupe_pyma
```

```python
# Checking the number of mice in the clean DataFrame.
num_nodupe_mouse = no_dupe_pyma['Mouse ID'].value_counts().count()
num_nodupe_mouse
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: mean, median, variance, standard deviation, and SEM of the tumor volume. 
grouped_pyma = no_dupe_pyma.groupby("Drug Regimen")
mean_tumor = grouped_pyma["Tumor Volume (mm3)"].mean()
median_tumor = grouped_pyma["Tumor Volume (mm3)"].median()
stddev_tumor = grouped_pyma["Tumor Volume (mm3)"].std()
variance_tumor = grouped_pyma["Tumor Volume (mm3)"].var()
SEM_tumor= grouped_pyma["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.
stat_summary_tumor = pd.DataFrame({"Mean Tumor Volume":mean_tumor,"Median Tumor Volume":median_tumor,"Tumor Volume Variance":variance_tumor,"Tumor Volume Std. Dev.":stddev_tumor,"Tumor Volume Std. Err.":SEM_tumor})
stat_summary_tumor
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
stat_summary2_tumor = no_dupe_pyma.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
stat_summary2_tumor
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timept_pyma = no_dupe_pyma.groupby("Mouse ID")["Timepoint"].max()
last_timept_pyma = last_timept_pyma.reset_index()


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_pyma = pd.merge(last_timept_pyma,no_dupe_pyma,on=["Mouse ID","Timepoint"])
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugreg_pyma = merge_pyma.loc[(merge_pyma["Drug Regimen"] == "Capomulin") | (merge_pyma["Drug Regimen"] == "Ramicane") | (merge_pyma["Drug Regimen"] == "Infubinol") | (merge_pyma["Drug Regimen"] == "Ceftamin")]
drug_list = drugreg_pyma["Drug Regimen"].unique()

# Create empty list to fill with tumor vol data (for plotting)
tumorvol_list = []
tumordrug_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drug_list:
    tumordrug_list.append(drug)
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumorvol_drug = drugreg_pyma.loc[drugreg_pyma["Drug Regimen"] == drug,"Tumor Volume (mm3)"]
    tumorvol_drug_df = pd.DataFrame(tumorvol_drug)
    
    # add subset 
    tumorvol_list.append(tumorvol_drug)
   
    # Determine outliers using upper and lower bounds
    quartiles = tumorvol_drug_df["Tumor Volume (mm3)"].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    tumorvol_out = tumorvol_drug_df.loc[(tumorvol_drug_df["Tumor Volume (mm3)"] > upper_bound) | (tumorvol_drug_df["Tumor Volume (mm3)"] < lower_bound),"Tumor Volume (mm3)"]
    print(f"{drug}'s potential outliers: {tumorvol_out}") 
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.set_xticklabels(tumordrug_list)
ax1.boxplot(tumorvol_list,0,'ro')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capo_pyma = no_dupe_pyma.loc[no_dupe_pyma["Drug Regimen"] == "Capomulin"] 
mouse_id = capo_pyma["Mouse ID"].sample()
mouse_tumor = capo_pyma.loc[no_dupe_pyma["Mouse ID"] == mouse_id.iloc[0],["Timepoint","Tumor Volume (mm3)"]]
mouse_tumor = mouse_tumor.set_index("Timepoint")
line_mouse_tumor = mouse_tumor.plot(kind="line", figsize=(8,5), legend=False, xlabel="Timepoint (days)", ylabel="Tumor Volume (mm3)",title=f"Capomulin treatment of mouse {mouse_id.iloc[0]}")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin = no_dupe_pyma.loc[no_dupe_pyma["Drug Regimen"] == "Capomulin"]

capomulin_tumor = capomulin.groupby("Mouse ID").mean()
plt.scatter(capomulin_tumor["Weight (g)"], capomulin_tumor["Tumor Volume (mm3)"], marker="o")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
corr = round(st.pearsonr(capomulin_tumor["Weight (g)"], capomulin_tumor["Tumor Volume (mm3)"])[0],2)
print(f"The correlation between mouse weight and the average tumor volume is {corr}")

pe_slope, pe_int, pe_r, pe_p, pe_std_err = stats.linregress(capomulin_tumor["Weight (g)"], capomulin_tumor["Tumor Volume (mm3)"])
pe_fit = pe_slope * capomulin_tumor["Weight (g)"] + pe_int
plt.scatter(capomulin_tumor["Weight (g)"], capomulin_tumor["Tumor Volume (mm3)"], marker="o")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.plot(capomulin_tumor["Weight (g)"],pe_fit,"--",color="r")
plt.show()
```
